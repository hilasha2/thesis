```python
def get_closest_samples(dif_dict):
    # sorted list by distance:
    sorted_distances = [(k, dif_dict[k]) for k in sorted(dif_dict, key=dif_dict.get, reverse=False)]
    for i in range(len(sorted_distances)):
        patient_number = dat.sample_metadata.loc[sorted_distances[i][0], 'patient_No']
        sorted_distances[i] = sorted_distances[i] + (patient_number,)
    return(sorted_distances)

get_closest_samples(dif_dict)
```
